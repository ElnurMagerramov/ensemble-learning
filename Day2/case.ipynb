```python
from sklearn.preprocessing import LabelEncoder
df[["Close Approach Date","Orbit Determination Date"]] = df[["Close Approach Date","Orbit Determination Date"]].apply(LabelEncoder().fit_transform)
```

```python
df.info()
```

```python
df.Hazardous.value_counts()
```

```python
y=df.Hazardous
X=df.drop("Hazardous",axis=1)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
```

```python
X_resampled, y_resampled = ros.fit_resample(X, y)
```

```python
y_resampled.value_counts()
```

```python
X_resampled.shape
```

```python
X_resampled.plot(kind="box",vert=False,figsize=(20,10))
```

```python
from sklearn.preprocessing import MinMaxScaler
X_resampled_copy=pd.DataFrame(MinMaxScaler().fit_transform(X_resampled),columns=X_resampled.columns).abs()
```

```python
X_resampled_copy.plot(kind="box")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X_resampled, y_resampled, test_size = 0.2, random_state = 0)
```

```python
from skopt import BayesSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
import xgboost as xgb
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
parameters = {
    "booster": ("gbtree", "gblinear", "dart"),
    "verbosity": (0, 1, 2, 3),
    "validate_parameters": (True, False),
    "disable_default_eval_metric": (True, False),
    "eta": list(np.arange(0, 1.1, 0.1)),
    "subsample": list(np.arange(0, 1.1, 0.1)),
}
```

```python
xgb_model = xgb.XGBClassifier(random_state=0)
```

```python
xgb_model.fit(X_train, y_train)
xgb_model.score(X_test, y_test)
```

```python
opt = BayesSearchCV(xgb.XGBClassifier(),parameters,n_iter=32,random_state=0)

opt.fit(X_train, y_train)
```

```python
print(" Results from Grid Search ")
print("\n The best estimator across ALL searched params:\n",opt.best_estimator_)
print("\n The best score across ALL searched params:\n",opt.best_score_)
print("\n The best parameters across ALL searched params:\n",opt.best_params_)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
print(classification_report(y_test, opt.predict(X_test)));
```
